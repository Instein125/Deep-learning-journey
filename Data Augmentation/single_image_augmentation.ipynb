```python
import keras.utils as image
from keras.preprocessing.image import ImageDataGenerator
```

```python
import os.path
from os import path

os.mkdir('/content/sample_data/train')
os.mkdir('/content/sample_data/augmented')
```

```python
img = image.load_img("/content/sample_data/train/cat.jpg", target_size = (200,200) )
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
img = image.img_to_array(img)
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)#we provide data in form of batches. here 1 represent we have only 1 image in our batch
```

```python
# generating augmented image
i = 0
for output in datagen.flow(input_batch, batch_size = 1, save_to_dir="/content/sample_data/augmented"):
  i+=1
  if (i==10):
    break

# we are using flow as we have single image.
# for number of images we use flow_from_directory
```
