```python
from keras.models import Model
```

Creating a non linear neural network which has input layer 1st dense layer of 128 neurons and 2nd dense layer of 64 neurons. We will have two output layers.

```python
# Creating model object
model = Model(inputs = x,outputs = [output1, output2])
```

```python
from keras.layers import *
```

```python
x = Input(shape=(3,))  #input has 3 columns
```

```python
# Creating hidden layer
hidden1 = Dense(128, activation = "relu")(x)  #here x is input to dense layer
hidden2 = Dense(64, activation = "relu")(hidden1)

#creating output layer
output1 = Dense(1,activation = "linear")(hidden2) #for linear regression
output2 = Dense(2, activation = "sigmoid")(hidden2) #for classification
```

```python
model.summary()
```

Visualizing the model

```python
from keras.utils import plot_model
plot_model(model, show_shapes = True)
```
